```python
def signal_generator():
    x1 = np.linspace(-np.pi, 0)
    print(x1)
    x2 = np.linspace(0, np.pi)
    y = x1 + 2 * x2
    return x1 + x2, y
```

```python
def signal_generator(argument_value):
    if -np.pi <= argument_value and argument_value < 0:
        return argument_value
    if 0 < argument_value <= np.pi:
        return 2 * argument_value
    if argument_value == 0:
        return 0
```

```python
domain = np.linspace(-np.pi, np.pi)
signal = np.array([signal_generator(signal_value) for signal_value in domain])
```

```python
def signal_as_fourier_series(domain, n=5):
    signal = np.pi / 4
    for k in range(1, n + 1):
        signal += ((-2 / np.pi * np.cos(k * domain) / k ** 2  if k % 2 == 1 else 0) + (3 * (-1) ** (k + 1) * np.sin(k * domain) / k))
    return signal   
```

```python
#signal_fourier_series = np.pi / 4 - 2 / np.pi * (np.cos(domain) / 1 ** 2 + np.cos(3 ))
signal_fourier_series = signal_as_fourier_series(domain, 1000)
```
